Write a program to reverse the order of elements in a given list. Implement this without
using any built-in functions or loops.

```r
reverseList <- function(lst) {
  if (length(lst) <= 1) {
    return(lst)
  } else {
    First <- lst[1]
    Remaining <- lst[-1]
    ReverseRemaining <- reverseList(Remaining)
    return(c(ReverseRemaining, First))
  }
}

lst <- c(1, 2, 3, 4, 5)

Reverse <- reverseList(lst)

cat("\nOriginal list: ", lst)
cat("\nReversed list: ", Reverse)
```

Create a scatter plot of the displacement (disp) versus miles per gallon in the mtcars dataset. Use different colors and sizes to represent the number of carburetors (carb). Add appropriate title, labels, and legend to the plot. Add a smooth line to show the trend of the relationship.

```r
# Load the mtcars dataset
data(mtcars)

# Create a scatterplot of disp versus wt
plot(mtcars$mpg, mtcars$disp, xlab = "Miles per gallon", ylab = "Displacement",
     main = "Scatterplot of Displacement vs Miles per gallon")

# Add a smooth line to show the trend
lines(lowess(mtcars$mpg, mtcars$disp), col = "blue", lwd = 2)

# Add a legend for the smooth line
legend("topright", legend = "Trendline", col = "blue", lty = 1, lwd = 2)
```

Create a histogram of the miles per gallon (mpg) in the mtcars dataset. Use different shades of blue to represent the frequency of each bin. Add appropriate title and labels to the plot. Calculate and display the mean and standard deviation of mpg on the plot.

```r
# Load the mtcars dataset
data(mtcars)

# Calculate mean and standard deviation of mpg
mpg_mean <- mean(mtcars$mpg)
mpg_sd <- sd(mtcars$mpg)

# Create the histogram with different shades of blue
hist(mtcars$mpg,
     breaks = 12,  # Adjust the number of bins as needed
     col = "lightblue", border = "darkblue",
     main = "Histogram of Miles per Gallon (mtcars)",
     xlab = "Miles per Gallon",
     ylab = "Frequency")

# Add mean and standard deviation lines
abline(v = mpg_mean, col = "red", lwd = 2)
abline(v = mpg_mean + mpg_sd, col = "orange", lty = 2, lwd = 2)
abline(v = mpg_mean - mpg_sd, col = "orange", lty = 2, lwd = 2)

# Add text for mean and standard deviation values
text(mpg_mean + 1, max(hist(mtcars$mpg)$counts) * 0.9, paste("Mean:", round(mpg_mean, 2)), col = "red")
text(mpg_mean + mpg_sd + 1, max(hist(mtcars$mpg)$counts) * 0.8, paste("SD:", round(mpg_sd, 2)), col = "orange")
text(mpg_mean - mpg_sd + 1, max(hist(mtcars$mpg)$counts) * 0.8, paste("SD:", round(mpg_sd, 2)), col = "orange")
```

Create a box plot of the horsepower (hp) in the mtcars dataset. Use different shapes to represent the number of gears (gear). Add appropriate title, labels, and legend to the plot. Identify and label any outliers on the plot.

```r
# Load the mtcars dataset
data(mtcars)

# Create a box plot with different shapes for each gear
boxplot(hp ~ gear, data = mtcars, col = c("red", "green", "blue"), pch = c(1, 2, 3),
        main = "Box Plot of Horsepower by Number of Gears",
        xlab = "Number of Gears", ylab = "Horsepower")

# Add a legend to the plot
legend("topright", legend = levels(factor(mtcars$gear)),
       col = c("red", "green", "blue"), pch = c(1, 2, 3), title = "Gears")

# Identify and label outliers
outliers <- boxplot.stats(mtcars$hp)$out
text(x = as.numeric(mtcars$gear[outliers]), y = outliers,
     labels = round(outliers, 2), pos = 4, col = "red")
```

Create a scatter plot of the displacement (disp) versus the weight (wt) in the mtcars dataset. Use different colors and sizes to represent the number of carburetors (carb). Add appropriate title, labels, and legend to the plot. Add a smooth line to show the trend of the relationship.

```r
# Load the mtcars dataset
data(mtcars)

# Define colors and sizes based on the number of carburetors
colors <- c("red", "green", "blue", "purple", "orange", "black", "brown", "cyan")
sizes <- c(1, 2, 3, 4, 5, 6, 7, 8)

# Create a scatterplot with colored and sized points, and a smooth line
plot(mtcars$wt, mtcars$disp, col = colors[mtcars$carb], pch = 16, cex = sizes[mtcars$carb],
     main = "Scatterplot of Displacement vs Weight", xlab = "Weight", ylab = "Displacement")

# Add a legend to the plot
legend("topright", legend = levels(factor(mtcars$carb)), col = colors, pch = 16, title = "Carburetors")

# Add a smooth line to show the trend
lines(lowess(mtcars$wt, mtcars$disp), col = "black", lwd = 2)

# Add a title
title(main = "Scatterplot of Displacement vs Weight with Carburetors", col.main = "blue", font.main = 4)
```
