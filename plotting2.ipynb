```r
# Load the mtcars dataset
data(mtcars)

# Create a scatterplot with different colors for each level of 'cyl'
plot(mtcars$disp, mtcars$mpg, xlab = "Displacement", ylab = "Miles per gallon",
     main = "Scatterplot of mpg and disp", col = mtcars$cyl)
# Add a legend
legend("topright", legend = factor(unique(mtcars$cyl)), fill = unique(mtcars$cyl), pch = 19)

# Save the plot as a high-resolution png file
png("scatterplot.png", width = 800, height = 600, res = 300)
dev.off()
```
