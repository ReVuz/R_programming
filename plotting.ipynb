```r
# Load the iris dataset, which is built-in in R
data(iris)

# Create a scatterplot of Sepal.Length and Petal.Length, and add labels and a title
# plot(iris$Sepal.Length, iris$Petal.Length, xlab = "Sepal Length", ylab = "Petal Length", main = "Scatterplot of Sepal Length vs Petal Length")

# Add different colors for each species using the col argument
# Use a vector of colors that matches the order of the levels of the species factor
plot(iris$Sepal.Length, iris$Petal.Length, xlab = "Sepal Length", ylab = "Petal Length", main = "Scatterplot of Sepal Length vs Petal Length", col = c("red", "blue", "green")[iris$Species])

# Add a legend to the top right corner using the legend function
# Use the levels of the species factor as the legend text
# Use the same vector of colors as before for the fill argument
# Use pch = 19 to make solid circles in the legend
legend("topright", legend = levels(iris$Species), fill = c("red", "blue", "green"), pch = 19)

# Save the plot as a high-resolution png file using the png function
png("scatterplot.png", width = 800, height = 600, res = 300)
dev.off()
```

```r
# Load the mtcars dataset, which is built-in in R
data(mtcars)

# Load the ggplot2 package for plotting
library(ggplot2)

# Create a scatterplot of mpg and disp, using different colors for cyl
ggplot(mtcars, aes(x = disp, y = mpg, color = factor(cyl))) +
  geom_point() +
  # Add a smooth line to show the trend, using loess method
  geom_smooth(method = "loess") +
  # Add labels and title
  labs(x = "Displacement", y = "Miles per gallon",
       color = "Number of cylinders", title = "Scatterplot of mpg and disp") +
  # Add a legend to the top right corner
  theme(legend.position = "topright")
```
